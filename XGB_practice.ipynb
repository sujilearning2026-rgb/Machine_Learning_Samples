## 1. Import Necessary Libraries

```python
import pandas as pd
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score
```

## 2. Data Collection

```python
breast_cancer_data = load_breast_cancer()
breast_cancer_data
breast_cancer_data_df = pd.DataFrame(data=breast_cancer_data.data, columns=breast_cancer_data.feature_names)
breast_cancer_data_df["Target"] = breast_cancer_data.target
breast_cancer_data_df
```

## 3. Data Understanding

```python
breast_cancer_data_df.isna().sum()
```

```python
breast_cancer_data_df.dtypes
```

## 4. Data Preparation

```python
X = breast_cancer_data_df.drop("Target", axis=1)

std_scaler  = StandardScaler()

scaled_X = pd.DataFrame(std_scaler.fit_transform(X))
scaled_X
```

```python
y = breast_cancer_data_df["Target"]
y
```

## 5. Model Building

```python
from sklearn.ensemble import GradientBoostingClassifier

gradient_boosting_classifier = GradientBoostingClassifier(max_depth=3, n_estimators=50)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=123, shuffle=True, stratify=y)

```

## 6. Model Training

```python
gradient_boosting_classifier.fit(X_train, y_train)
```

## 7. Model Testing || 8. Model Evaluation

```python
y_pred_train = gradient_boosting_classifier.predict(X_train)

print(confusion_matrix(y_train, y_pred_train))
print(accuracy_score(y_train, y_pred_train))
print(precision_score(y_train, y_pred_train))
print(recall_score(y_train, y_pred_train))
```

```python
y_pred_test = gradient_boosting_classifier.predict(X_test)

print(confusion_matrix(y_test, y_pred_test))
print(accuracy_score(y_test, y_pred_test))
print(precision_score(y_test, y_pred_test))
print(recall_score(y_test, y_pred_test))
```
