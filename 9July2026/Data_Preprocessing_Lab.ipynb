## 4. Data Preparation

#### 4.1 Data Cleaning - drop, fillna with Mean, Median, Mode Imputation, Changing Data Types

```python
weather_data.head()
```

```python
#drop unnecessary columns like Unnamed. Drop Temp C as Temp column is already available
weather_data.drop(labels=["Unnamed: 0", "Temp C"], axis=1, inplace=True)
```

```python
weather_data
```

```python
#Unwanted columns (Unnamed and Temp C) are removed as mentioned.
```

```python
#check for null values
weather_data.isna().sum()
```

```python
#38 null entries are available on Ozone feature, Since this shows 25% of the data on actual dataset are null, this column is removed
```

```python
del weather_data["Ozone"]
```

```python
weather_data.dtypes
```

```python
weather_data.isna().sum()
```

```python
# Solar.R feature contains 7 null entries and this can be filled with mean data
```

```python
weather_data.groupby("Weather")["Solar.R"].mean().round()
```

```python
#transform
weather_data["Solar.R"] = weather_data.groupby("Weather")["Solar.R"].transform(lambda x: x.fillna(x.mean().round()))
weather_data
```

```python
weather_data.isna().sum()
```

```python
#important - if the output feature (Weather column) is NaN and Solar.R is also NaN, delete those rows

weather_data.dropna(inplace=True)
```

```python
weather_data.isna().sum()
```

```python
#remove null entries
#fill with mean median mode imputation
#change the data type
#get unique values of Month column

weather_data["Month"]. unique()
```

```python
#change the value of Month feature from 'May' to 5
import warnings
warnings.filterwarnings(action="ignore")
weather_data["Month"].replace(to_replace="May", value=5, inplace=True)
```

```python
#change the data type of Month column to int

weather_data["Month"] = weather_data["Month"].astype("int")
weather_data.dtypes
```

```python
#after data cleaning, input features data type is in numerics and output feature is in object, it has to be addressed
```

#### 4.2 Data Transformation 

```python
#keep a copy before applying transformation as the data is cleaned
weather_data_copy_01 = weather_data.copy()
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

le_encoder = LabelEncoder()

#fit and transform Weather feature (S = 2, C = 0, PS = 1)
weather_data_copy_01["Weather"] = le_encoder.fit_transform(weather_data_copy_01["Weather"])
```

```python
weather_data_copy_01
```

#### OneHotEncoder usage
