## 1. Import Necessary Libraries

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split, KFold, LeaveOneOut, cross_val_score
```

## 2. Data Collection

```python
claimants_data = pd.read_csv("claimants_data.csv")
claimants_data
```

## 4. Data Preparation

#### 4.1 Data Cleaning

```python
claimants_data.dropna(inplace=True)
claimants_data
claimants_data.isna().sum()
```

#### 4.2 Separate Input and Output

```python
X = claimants_data.drop(labels=["CASENUM", "ATTORNEY"], axis=1)
```

```python
X
```

```python
y= claimants_data["ATTORNEY"]
y
```

## 5. Model Building

#### Model Validation Techniques

##### 5.1 Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, shuffle=True, random_state=42)
X_train
```

```python
X_test
```

##### 5.2 K-Fold Cross Validation

```python
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
kfold.n_splits
```

```python
logistics_model = LogisticRegression()
```

```python
cv_scores = cross_val_score(estimator=logistics_model, X=X, y=y,cv=kfold)
print(cv_scores)
print("Overall Accuracy:", cv_scores.mean().round(2))
print("Overall Deviation:", cv_scores.std().round(2))
```

#### 5.3 LOOCV - Leave One Out Cross Validation

```python
l_cv = LeaveOneOut()
```

```python
cv_scores = cross_val_score(estimator=logistics_model, X=X, y=y,cv=l_cv)
print(cv_scores)
print("Overall Accuracy:", cv_scores.mean().round(2))
print("Overall Deviation:", cv_scores.std().round(2))
```
